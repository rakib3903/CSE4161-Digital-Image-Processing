```python
import numpy as np
import cv2
import matplotlib.pyplot as  plt

```

```python
img = cv2.imread('cat.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
grayImage = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
# (a) Perform thebrightness enhancement of a specific range of gray levels and observe its result

t1 = grayImage.copy()
t2 = grayImage.copy()
for i in range(grayImage.shape[0]):
    for j in range(grayImage.shape[1]):
        if 150 <= grayImage[i][j] <= 230:
            t1[i][j] = 255
            t2[i][j] = 255
        else :
            t1[i][j] = 0


plt.figure(figsize=(8, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.axis('off')
plt.imshow(grayImage, cmap='gray')

# Image enhancement within boundary range with 255 and 0 for out of boundary
plt.subplot(1, 3, 2)
plt.title('Image Enhancement 1')
plt.axis('off')
plt.imshow(t1, cmap='gray')


# Image enhancement within boundary range with 255 and unalterd for out of boundary
plt.subplot(1, 3, 3)
plt.title('Image Enhancement 2')
plt.axis('off')
plt.imshow(t2, cmap='gray')
```

```python
# (b) Differentiate the results of power law transform and inverse logarithmic transform

PowerLawImg = np.power(grayImage / 255, 4.0) * 255

c = 255 / np.log(1 + 255)
InverseLawImg = np.exp(grayImage / c) - 1

plt.figure(figsize=(8, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.axis('off')
plt.imshow(grayImage, cmap='gray')

# Power Law Transformation
plt.subplot(1, 3, 2)
plt.title('Power Law Tranformation')
plt.axis('off')
plt.imshow(PowerLawImg, cmap='gray')


# Inverse Law Transformation
plt.subplot(1, 3, 3)
plt.title('Inverse Law Tranformation')
plt.axis('off')
plt.imshow(InverseLawImg, cmap='gray')
```

```python
# (c) Find the difference image between original and the image obtained by last three bits (MSB)

bitPlaneImg = grayImage & 224  

plt.figure(figsize=(8, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.axis('off')
plt.imshow(grayImage, cmap='gray')

# Last three MSB bit plane image
plt.subplot(1, 3, 2)
plt.title('MSB-3 bit plane image')
plt.axis('off')
plt.imshow(bitPlaneImg, cmap='gray')


# Difference between original and msb 3 bit plane image
plt.subplot(1, 3, 3)
plt.title('original - bitPlane image')
plt.axis('off')
plt.imshow(InverseLawImg - bitPlaneImg, cmap='gray')

```
