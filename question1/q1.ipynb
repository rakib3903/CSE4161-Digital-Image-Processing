```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("cat.jpg")
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grayImage, cmap='gray')

print(img.shape)
```

```python
# saptial resulation by half operation
img1 = img.copy()
temp = []
for i in range(6):
    temp.append(img1)
    img1 = img1[::2, ::2] 


plt.figure(figsize=(10, 6))
for i in range(6):
    plt.title(f"{temp[i].shape[0]} x {temp[i].shape[1]}")
    plt.subplot(2, 3, i+1)
    plt.imshow(temp[i], cmap='gray')
```

```python
# image historgram and thresholding

histogram = np.zeros(256, dtype = int)

for i in range(grayImage.shape[0]):
    for j in range(grayImage.shape[1]):
        histogram[grayImage[i][j]] += 1

threshold = grayImage.copy()
threshold[threshold < 100] = 0
threshold[threshold >= 100] = 255

plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.bar(range(256), histogram, color='gray')


plt.subplot(1, 2, 2)
plt.axis('off')
plt.title("thresholding")
plt.imshow(threshold, cmap='gray')
```
